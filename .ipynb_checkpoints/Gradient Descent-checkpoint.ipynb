```python
import numpy as np
from sympy import *
from sympy.abc import x
```

```python
expr =exp(x)-log(x)
dy_dx= diff(expr,x)
xi =1.5
eta = 0.01
```

```python
for i in range(10):
    
    new_x = xi-eta*dy_dx.subs([(x,xi)])
    print()
    print("xi=",new_x )
    print("f(xi) = ",expr.subs([(x,new_x)]))
    xi= new_x
```

```python
expr =x**3-6*x**2+4*x+2
dy_dx= diff(expr,x)
xi =1.5
eta = 0.01
```

```python
for i in range(50):
    
    new_x = xi-eta*dy_dx.subs([(x,xi)])
    print()
    print("xi=",new_x )
    print("f(xi) = ",expr.subs([(x,new_x)]))
    xi= new_x
```

```python
xi = np.array([4,6,7,8,1,9])
yi = np.array([17,23,26,29,8,32])

eta = 0.08
b0 = 0.4
w0=-0.6
ycap = xi*w0+b0

for i in range(1,1000):
    
    db = -(1/len(xi))*np.sum(yi-ycap)
    dw = -(1/len(xi))*np.sum((yi-ycap)*xi)
    new_b= b0-eta*db
    new_w = w0-eta*dw
    b0 = new_b
    w0 = new_w
    ycap = xi*w0+b0
    print("L = ",(1/2*len(xi))*np.sum(yi-ycap))
    print("w0 = ",w0,"b0 = ",b0)





```
